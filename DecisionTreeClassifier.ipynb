## **Problem Statement:**

1.   Create the Decision Tree classifier for the dataset and visualize it graphically.
2.   The purpose is if we feed any new data to this classifier, it would be able to 
predict the right class accordingly. 




**STEP 1: Importing Libraries**

The following script imports required libraries:

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
%matplotlib inline
```

**STEP 2: Loading the Dataset**

Since our file is in CSV format, we will use panda's read_csv method to read our CSV data file:

```python
data=pd.read_csv('/content/Iris.csv')
```

```python
data
```

**STEP 3: Preparing the Data**

Now, We will explore the data to gain insights about the data.

```python
data=data.drop("Id",axis=1) #we don't need ID column parameter to determine the Output. 
```

Execute the following command to see the number of rows and columns in our dataset

```python
data.shape #view the dimensions of the new data
```

```python
data.info() #view the summary of the Dataset
```

```python
data.isnull().sum() #checking missing values in parameters
```

We can see there are No missing values in the Dataset.

**STEP 4: Declare Feature Vector and Target Variable**

The X variable contains all the columns from the dataset, except the "Species" column, which is the target variable. The y variable contains the values from the "Species" column. The X variable is our attribute set and y variable contains corresponding values.

```python
X = data.drop('Species', axis=1)
y = data['Species']
```

**STEP 5 : Spliting data into Training and Testing sets**

The final preprocessing step is to divide our data into training and test sets. The model_selection library of Scikit-Learn contains train_test_split method, which we'll use to randomly split the data into training and testing sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 100)
```

```python
X_train
```

```python
y_train
```

**STEP 6: Train the Algorithm**

The final step is to train the decision tree algorithm on this data and make predictions. Scikit-Learn contains the tree library, which contains built-in classes/methods for various decision tree algorithms. Since we are going to perform a classification task here, we will use the DecisionTreeClassifier class for this example. The fit method of this class is called to train the algorithm on the training data.

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

Now that our classifier has been trained, let's make predictions on the test data. To make predictions, the predict method of the DecisionTreeClassifier class is used.

```python
y_pred = classifier.predict(X_test)
```

Predicting the Output by the given Dataset whether it's predicting Right class or not. 

(Note that, here the classes are the types of outcome i.e 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica')

```python
classifier.predict([[4.3,3,1.1,0.1]])
```

**STEP 7: Evaluating the Algorithm (ACCURACY)** 

At this point we have trained our algorithm and made some predictions. Now we'll see how accurate our algorithm is. Scikit-Learn's metrics library contains the Accuracy method that can be used to calculate this metric for us.

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

**STEP 8: Plot the Decision tree** 

plot_tree is used to plot the Decision tree.The figsize argument of plt.figure will control the size of the rendering.

```python
plt.figure(figsize=(12,8))
from sklearn import tree
tree.plot_tree(classifier.fit(X_train, y_train))
```

**STEP 9: Visualizing the dataset graphically**

export_graphviz function converts decision tree classifier into dot file and pydotplus convert this dot file to png or displayable form on IDE.

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus
dot_data = StringIO()
export_graphviz(classifier, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names =['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=['Iris-setosa','Iris-versicolor','Iris-virginica'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Iris.png')
Image(graph.create_png())
```

**STEP 10: Predicting the Right class accordingly**

As the problem statement states that the purpose is if we feed any new data to this classifier, it would be able to 
predict the right class accordingly.As we have already predicted the output.Now we are going to add new data and predict the right class.

```python
classifier.predict([[7.5,3,6.3,1.8]])
```
